```python
# delete duplicates by column 'positionId' and delete all the others except the first one
df_duplicates = df.drop_duplicates(subset = 'positionId', keep = 'first') 
df_duplicates.head(5)
```

```python
df_duplicates.info()
```

```python
# Create a function to extract bottom and top salary
def cut_word(word, method):
    position = word.find('-')
    length = len(word)
    # regular case, e.g. 10k-15k
    if position != -1:
        bottomSalary = word[:position-1]
        topSalary = word[position+1:length-1]
    # special case, e.g. 15k以上
    else:
        bottomSalary = word[:word.upper().find('K')]
        topSalary = bottomSalary
    if method == 'bottom':
        return bottomSalary
    else:
        return topSalary
```

```python
df_duplicates['bottomSalary'] = df_duplicates.salary.apply(cut_word, method = 'bottom')
df_duplicates['topSalary'] = df_duplicates.salary.apply(cut_word, method = 'top')
```

```python
# To prevent the above warning, we should create correct Series and then insert into data frame
# Method 1. We can use insert() 
# bottomSalary = df_duplicates.salary.apply(cut_word, method = 'bottom')
# df_duplicates.insert(df_duplicates.shape[1], 'bottomSalary', bottomSalary) # dataframe.shape[1] is the number of columns

# Method2. We can use assign()
df_duplicates = df_duplicates.assign(bottomSalary = df_duplicates.salary.apply(cut_word, method = 'bottom'),
                    topSalary = df_duplicates.salary.apply(cut_word, method = 'top'))

df_duplicates.head()

# Note: we should use one assign() function to add two columns. 
# If we seperate it into two lines of commands, the latter one will overwrite the former which is not what we want.

# To calculate bottom salary, we can also use lambda function. The same to calculate top salary.
# bs = df_duplicates.apply(lambda x: cut_word(x.salary, 'bottom'), axis = 1)
```

```python
pd.options.mode.chained_assignment = None # disable "SettingWithCopyWarning" warning
df_duplicates['bottomSalary'] = df_duplicates['bottomSalary'].astype('int')
```

```python
df_duplicates['topSalary'] = df_duplicates['topSalary'].astype('int')
```

```python
df_duplicates = df_duplicates.assign(avgSalary = df_duplicates.apply(lambda x: (x.bottomSalary+x.topSalary)/2, axis=1))
df_duplicates.head()
```

```python
# after the above step, the returned result is still a Series. We have to add "str" before using "replace"
word = df_clean.positionLables.str[1:-1].str.replace(' ', '')
```

```python
# Remove null value
# Split labels
word.dropna().str.split(',').apply(pd.value_counts)
```

```python
df_word = word.dropna().str.split(',').apply(pd.value_counts)
df_word.unstack()
```

```python
# Delete null value and reset to DataFrame
# level_0 is label and level_1 is the index of df_word which we can think it is mapping to a position
# "0" column is the occurance of the label in the position. Since we didn't name it, it shows the default name "0"

df_word.unstack().dropna().reset_index()
```

```python
# Group by label and count the occurance
df_word_counts = df_word.unstack().dropna().reset_index().groupby('level_0').count()
```

```python
from wordcloud import WordCloud
df_word_counts.index = df_word_counts.index.str.replace("'", "")
wordcloud = WordCloud(font_path = "C:\Windows\Fonts\msyh.ttf",
                     width = 900, height = 400,
                     background_color = 'white')
f, axs = plt.subplots(figsize = (15, 15))
wordcloud.fit_words(df_word_counts.level_1)
axs = plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
