```python
df_dedup = df.drop_duplicates(subset = 'positionId', keep = 'first')
```

```python
df_dedup.info()
```

```python
import re

def cut_word(word, method, bottomSalary = 0, topSalary = 0):
    try:
        bottomSalary = re.findall("\d+", word)[0]
        topSalary = re.findall("\d+", word)[1]
    except IndexError:
        topSalary = bottomSalary
    if method == 'top':
        return topSalary
    else:
        return bottomSalary
```

```python
df_dedup = df_dedup.assign(bottomSalary = df_dedup.salary.apply(cut_word, method ='bottom'),
                          topSalary = df_dedup.salary.apply(cut_word, method = 'top'))
df_dedup.head()
```

```python
df_dedup.is_copy = False
df_dedup['bottomSalary'] = df_dedup['bottomSalary'].astype(int)
df_dedup['topSalary'] = df_dedup['topSalary'].astype(int)
```

```python
df_dedup.info()
```

```python
df_dedup['avgSalary'] = df_dedup.apply(lambda x: (x.bottomSalary + x.topSalary)/2, axis = 1)
```

```python
df_dedup.head()
```

```python
df_clean = df_dedup[['city', 'companyShortName', 'companySize', 'education', 
                          'positionName', 'positionLables', 'workYear', 'avgSalary']]
df_clean.head()
```
